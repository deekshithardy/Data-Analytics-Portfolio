# Data Visualization

The following visualizations provide insights into sales performance, regional distribution, profit distribution, and state-wise sales.

```python
# Sales by Category

category_sales = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(8,5))
category_sales.plot(kind="bar")

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)

plt.tight_layout()
plt.savefig("../screenshots/sales_by_category.png", dpi=300)

plt.show()
```

```python
# Sales Distribution by Category

category_sales.plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(7,7)
)

plt.title("Sales Distribution by Category")
plt.ylabel("")

plt.tight_layout()
plt.savefig("../screenshots/category_distribution.png", dpi=300)

plt.show()
```

```python
# Profit Distribution

plt.figure(figsize=(8,5))

plt.hist(df["Profit"], bins=20)

plt.title("Profit Distribution")
plt.xlabel("Profit")
plt.ylabel("Frequency")

plt.tight_layout()
plt.savefig("../screenshots/profit_distribution.png", dpi=300)

plt.show()
```
